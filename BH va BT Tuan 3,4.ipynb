```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.subplot(2,2,1)
plt.plot(x, -x**2, color = 'g')
plt.subplot(2,2,2)
plt.plot(x, -x**3, color = 'r')
plt.subplot(2,2,3)
plt.plot(x, -x**3+15, color = 'b')
plt.subplot(2,2,4)
plt.plot(x, -x, color = 'y')
plt.show()
```
